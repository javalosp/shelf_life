# Descriptive Statistical Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
input_file = "data/experimental_data_origin.xlsx"
output_dir = "outputs/description"
os.makedirs(output_dir, exist_ok=True)
```

```python
moisture = pd.read_excel(input_file, sheet_name="moisture")
oxidation = pd.read_excel(input_file, sheet_name="oxidation")

print("Moisture:")
display(moisture.head())

print("Oxidation:")
display(oxidation.head())
```

```python
moisture[["water_activity", "moisture_content"]].describe()
```

```python
oxidation[["set_point_t", "ip_h"]].describe()
```

```python
# Moisture: water_activity vs moisture_content
plt.figure(figsize=(8,6))
sns.scatterplot(data=moisture, x="water_activity", y="moisture_content", hue="category", alpha=0.8, s=20)
plt.title("Water Activity vs Moisture Content", fontweight="bold")
plt.xlabel("Water Activity")
plt.ylabel("Moisture Content")
plt.legend(title="Category", loc="upper left", bbox_to_anchor=(0.02, 0.98))
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "moisture_scatter.png"), dpi=300)
plt.show()
plt.close()
```

```python
# Oxidation: set_point_t vs IP
plt.figure(figsize=(8,6))
sns.scatterplot(data=oxidation, x="set_point_t", y="ip_h", hue="category", alpha=0.8, s=20)
plt.title("Temperature vs Induction Period", fontweight="bold")
plt.xlabel("Temperature (°C)")
plt.ylabel("Induction Period (hours)")
plt.legend(title="Category", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "oxidation_scatter.png"), dpi=300)
plt.show()
plt.close()
```
